**1. (Plotnine - Scatter Plot)**
Write a Python script using `plotnine` to create a **scatter plot** that visualizes the relationship between 'sepal_length' and 'sepal_width' from the `iris` dataset. Make sure to add a title and axis labels.

````python
# Write your code here
from plotnine import *
plot = (
    ggplot(df, aes(x='sepal_length', y='sepal_width'))
    + geom_point()
    + labs(title='Sepal Length vs. Sepal Width', x='Sepal Length (cm)', y='Sepal Width (cm)')
)

plot
````

**3. (Plotnine - Box Plot)**
Generate a **box plot** using `plotnine` to compare the distribution of 'sepal_length' across different species in the `iris` dataset.

````python
# Write your code here
plot = (
     ggplot(df, aes(x='species', y='sepal_length', fill='species'))
    + geom_boxplot()
    + labs(title='Distribution of Sepal Length by Species', x='Species', y='Sepal Length (cm)')
    + scale_fill_manual(values=['skyblue', 'lightgreen', 'salmon'])
)

plot
````

**4. (Plotly - Line Chart)**
Use `plotly.express` to create a **line chart** showing the sales trend over five years. The dataset should contain 'Year' and 'Sales' columns.

NB: Find the dataset online

````python
import seaborn as sns
import pandas as pd
import plotly.express as px

flights = sns.load_dataset("flights")

sales_data = flights.pivot_table(index='year', values='passengers', aggfunc='sum').reset_index()
sales_data = sales_data.rename(columns={'passengers': 'Sales'})
sales_data = sales_data.rename(columns={'year': 'Year'})
sales_data = sales_data.tail(5)

fig = px.line(sales_data, x='Year', y='Sales',
              title='Total Flight Sale Over Five Years',
              markers=True)

fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Total Sales',
    xaxis=dict(tickmode='linear'),
    yaxis=dict(tickformat=',.0f'),
)

fig.show()
````

**11. (Bokeh - Bar Chart)**
Create a **vertical bar chart** using Bokeh, where x values are `['A', 'B', 'C', 'D']` and y values are `[3, 7, 5, 2]`. Choose different colors for each bar.

````python
# Write your code here
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

output_notebook()

x = ['A', 'B', 'C', 'D']
y = [3, 7, 5, 2]

# Directly specify colors for each column
colors = ['pink', 'lightgreen', 'skyblue', 'orange']

source = ColumnDataSource(data=dict(x=x, y=y, colors=colors))

p = figure(x_range=x, title="Vertical Bar Chart", x_axis_label="Categories", y_axis_label="Values")

p.vbar(x='x', top='y', width=0.8, color='colors', source=source, legend_field='x')

show(p)
````

**12. (Bokeh - Interactive Legends)**
Modify a **Bokeh line plot** so that clicking on the legend hides the corresponding line.

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend, LegendItem
from bokeh.palettes import Category10

output_notebook()

x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [4, 5, 5, 7, 2]

p = figure(title="Interactive Legend Example", x_axis_label='X-Axis', y_axis_label='Y-Axis')

line1 = p.line(x, y1, legend_label="Line 1", line_color=Category10[3][0], line_width=2)
line2 = p.line(x, y2, legend_label="Line 2", line_color=Category10[3][1], line_width=2)
line3 = p.line(x, y3, legend_label="Line 3", line_color=Category10[3][2], line_width=2)

legend = Legend(items=[
    LegendItem(label="Line 1", renderers=[line1]),
    LegendItem(label="Line 2", renderers=[line2]),
    LegendItem(label="Line 3", renderers=[line3])
], click_policy="mute")

p.add_layout(legend, 'right')

show(p)
````

**14. (Bokeh - Scatter Plot)**
Write a Python script using `bokeh.plotting` to create a **scatter plot** with random data points, setting colors randomly using `bokeh.palettes`. Use at least 100 points.

````python
# Write your code here
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Category20

output_notebook()

num_points = 100
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 10

colors = [Category20[20][i % 20] for i in range(num_points)]

p = figure(title="Scatter Plot with Random Colors", x_axis_label='X-Axis', y_axis_label='Y-Axis')

p.circle(x, y, size=10, color=colors)

show(p)
````

**15. (Custom Visualization)**
Using **any** of the three libraries (`plotnine`, `plotly`, or `bokeh`), create a **custom visualization** of your choice with at least **two features** (e.g., different colors, interactive elements, multiple plots).

````python
import plotly.express as px
import numpy as np
import pandas as pd

np.random.seed(42)
num_points = 100
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 10
categories = np.random.choice(['Category A', 'Category B', 'Category C'], size=num_points)
df = pd.DataFrame({'x': x, 'y': y, 'category': categories})

fig = px.scatter(
    df,
    x='x',
    y='y',
    color='category',
    title="Interactive Scatter Plot with Categories",
    color_discrete_map={
        'Category A': 'pink',
        'Category B': 'lightgreen',
        'Category C': 'skyblue'
    },
    hover_data={'x': True, 'y': True, 'category': True}
)

fig.show()
````
